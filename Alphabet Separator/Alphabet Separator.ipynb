```python
import cv2
import numpy as np
#import matplotlib as plt
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('OCR.jpg' , cv2.IMREAD_GRAYSCALE)
plt.imshow(image,cmap = 'gray')
```

```python
thresh=cv2.adaptiveThreshold(image,100,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY_INV,5,4)
plt.imshow(thresh,cmap = 'gray')
```

```python
contours,_=cv2.findContours(thresh,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
```

```python
i=0
for c in contours:
    #if len(c)<100:
        x, y, w, h = cv2.boundingRect(c)
        #image[y:y+h,x:x+w]=255
        cv2.rectangle(image,(x,y),(x+w,y+h),(166,166,166),4)
        alphabet=image[y:y+h,x:x+w]
        i+=1
        cv2.imwrite("outputs/"f"{i}.jpg",alphabet)
        
plt.imshow(image,cmap='gray')
```
